## Data Normalization

```python
df[["length", "width", "height"]].head(10)
```

### Methods of Normalizing Data
**1. Simple Feature Scalling**<br>
Xnew = Xold/Xmax<br>
<br>
**2. Min-Max**<br>
Xnew = (Xold-Xmin)/(Xmax-Xmin)<br>
<br>
**3. Z-score**<br>
Xnew = (Xold-mean)/std<br>

```python
# Simple feature scalling
df6 = df
df6["length"] = df6["length"]/df6["length"].max()
df6[["length", "width", "height"]].head(10)
```

```python
# Min-Max
df7 = df
df7["length"] = (df7["length"] - df7["length"].min())/(df7["length"].max() - df7["length"].min())
df7[["length", "width", "height"]].head(10)
```

```python
# Z-score
df8 = df
df8["length"] = (df8["length"] - df8["length"].mean())/df8["length"].std()
df8[["length", "width", "height"]].head(10)
```

## ANOVA (Analysis of Variance)
used to find the correlation between different groups of a categorical variable

**F-test score**<br>
variation between sample group means divided by variation within sample group<br>
**p-value**<br>
confidence degree<br>

```python
df14 = df11[["make", "price"]]
df15 = df14.groupby(["make"])
```

```python
import scipy.stats as stats
```

```python
# ANOVA between Honda and Subaru
df16 = stats.f_oneway(df15.get_group("honda")["price"], df15.get_group("subaru")["price"])
df16

# F-test score = 0.19744030127462606 (sangat kecil; < 1)
# p-value = 0.6609478240622193 (> 0.05)
```

```python
# ANOVA between Honda and Jaguar
df17 = stats.f_oneway(df15.get_group("honda")["price"], df15.get_group("jaguar")["price"])
df17

# F-test score = 400.925870564337 (sangat besar, strong correlation)
# p-value = 0.6609478240622193 (> 0.05)
```

## Pearson Correlation

1. Correlation Coefficient<br>
close to +1: large positive relationship<br>
close to -1: large negative relationship<br>
close to 0: no relationship<br>
<br>
2. P-Value<br>
p-value < 0.001: strong certainty in the result<br>
p-value < 0.05: moderate certainty in the result<br>
p-value < 0.1: weak certainty in the result<br>
p-value > 0.1: no certainty in the result<br>
<br>
**Strong Correlation**<br>
Correlation coefficient is close to 1 or -1 AND P-value is < 0.001

```python
# Correlation between engine size and price
pearson_coef, p_value = stats.pearsonr(df11["engine_size"], df11["price"])
print("Pearson Correlation =", pearson_coef)
print("P-value =", p_value)
```

```python
# Correlation between highway miles per galon and price
pearson_coef, p_value = stats.pearsonr(df11["highway_mpg"], df11["price"])
print("Pearson Correlation =", pearson_coef)
print("P-value =", p_value)
```
